```python
from rdflib import Graph, RDF, Namespace, Literal, URIRef

def serialize():
    print g.serialize(format='turtle')

def save(filename):
    with open(filename, 'w') as f:
        g.serialize(f, format='turtle')

g = Graph()
EX = Namespace('http://example.com/formula_one/')
g.bind('ex',EX)
owl = Namespace('http://www.w3.org/2002/07/owl#')
g.bind('owl',owl)
all_grandprixes = grandprix2['GP of']
g.add(( EX.in_country, RDF.type, owl.ObjectProperty) )
g.add(( EX.in_city, RDF.type, owl.ObjectProperty) )
g.add(( EX.name_circuit, RDF.type, owl.ObjectProperty) )
g.add(( EX.on_date, RDF.type, owl.DatatypeProperty) )
g.add(( EX.length_circuit, RDF.type, owl.DatatypeProperty) )
g.add(( EX.number_of_labs, RDF.type, owl.DatatypeProperty) )
g.add(( EX.total_distance, RDF.type, owl.DatatypeProperty) )
for grandprix in all_grandprixes:
    grandprix_info = grandprix2.loc[grandprix2['GP of']== grandprix]
    grandprix = grandprix.replace(" ", "_")
    country = grandprix_info['Country'].item().replace(" ", "_")
    city = grandprix_info['City'].item().replace(" ", "_")
    date = grandprix_info.date.dt.date.item()
    name = grandprix_info['name'].item().replace(" ", "_")
    length = grandprix_info['length [m]'].item()
    laps = grandprix_info['laps'].item()
    distance = grandprix_info['Total distance [km]'].item()
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), RDF.type, EX.Grandprix) )
    g.add(( URIRef('http://example.com/formula_one/' + country), RDF.type, EX.Country) )
    g.add(( URIRef('http://example.com/formula_one/' + city), RDF.type, EX.City) )
    g.add(( URIRef('http://example.com/formula_one/' + name), RDF.type, EX.Circuit) )
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.in_country, URIRef('http://example.com/formula_one/' + country)) )
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.in_city,  URIRef('http://example.com/formula_one/' + city)) )
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.name_circuit,  URIRef('http://example.com/formula_one/' + name )) )
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.on_date, Literal(date) ))
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.length_circuit, Literal(length) ))
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.number_of_labs, Literal(laps) ))
    g.add(( URIRef('http://example.com/formula_one/' + grandprix), EX.total_distance, Literal(int(distance)) ))
    
   
serialize()    
save('grandprix_info.ttl')
```

```python
from rdflib import Graph, RDF, Namespace, Literal, URIRef
from SPARQLWrapper import SPARQLWrapper, JSON

g = Graph()

EX = Namespace('http://example.com/formula_one/')
g.bind('ex',EX)
GEO = Namespace('http://www.opengis.net/ont/geosparql#')
g.bind('geo',GEO)
all_cities = grandprix2['City for Coordinates']
g.add(( GEO.longitude, RDF.type, owl.DatatypeProperty) )
g.add(( GEO.latitude, RDF.type, owl.DatatypeProperty) )
for city in all_cities:
    info_city = grandprix2.loc[grandprix2['City for Coordinates']== city]
    real_city = info_city['City'].item().replace(" ", "_")
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery("""
        PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
        PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
        PREFIX dbr: <http://dbpedia.org/resource/>
        PREFIX db: <http://dbpedia.org/>
        PREFIX dbo: <http://dbpedia.org/ontology/>
        SELECT * WHERE {
        dbr:"""+ city.replace(" ", "_") + """ geo:long ?long; geo:lat ?lat .
        } 
    """)
    sparql.setReturnFormat(JSON)
    resultt = sparql.query().convert()
    lat = resultt['results']['bindings'][0]['lat']['value']
    longi = resultt['results']['bindings'][0]['long']['value']
    g.add(( URIRef('http://example.com/formula_one/' + real_city), RDF.type, EX.City ))
    g.add(( URIRef('http://example.com/formula_one/' + real_city), GEO.longitude, Literal(longi) ))
    g.add(( URIRef('http://example.com/formula_one/' + real_city), GEO.latitude, Literal(lat) ))

serialize()
save('city_info.ttl')
```

```python
def shortenl(uri):
    if isinstance(uri, Literal):
        return uri
    components = g.namespace_manager.compute_qname(uri)
    return "%s"%(components[2])

g = Graph()

EX = Namespace('http://example.com/formula_one/')
g.bind('ex',EX)

DBO = Namespace('http://dbpedia.org/ontology/')
g.bind('dbo', DBO)
sparql = SPARQLWrapper("http://localhost:5820/Eind/query")
sparql.setQuery("""
PREFIX ex: <http://example.com/formula_one/>
SELECT ?o WHERE {?s ex:in_country ?o }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
g.add(( DBO.currency, RDF.type, owl.ObjectProperty) )
g.add(( DBO.language, RDF.type, owl.ObjectProperty) )
g.add(( DBO.abstract, RDF.type, owl.DatatypeProperty) )
for result in results["results"]["bindings"]:
    country = (shortenl(result["o"]["value"])) # Shortenl changes for example http://example.com/kad2017/Amserdam into Amsterdam
    g.add( (URIRef(result["o"]["value"]), RDF.type, EX.Country) )
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery(
        """ PREFIX dbr: <http://dbpedia.org/resource/>
            PREFIX dbo: <http://dbpedia.org/ontology/>
        
            SELECT ?abstract ?currency ?language WHERE {{dbr:"""+country+""" dbo:abstract ?abstract FILTER langMatches( lang(?abstract), "en" )} UNION {dbr:"""+country+""" dbo:currency ?currency} UNION {dbr:"""+country+""" dbo:officialLanguage ?language FILTER NOT EXISTS{?language rdf:type dbo:Organisation}}}""")
    sparql.setReturnFormat(JSON)
    results2 = sparql.query().convert()
    for result2 in results2["results"]["bindings"]:
        for key, value in result2.items():
            if key == 'abstract':
                g.add( (URIRef(result["o"]["value"]), DBO.abstract, Literal(value['value'])))
            elif key == 'currency':
                g.add( (URIRef(result["o"]["value"]), DBO.currency, URIRef(value['value'])))
                g.add( (URIRef(value['value']), RDF.type, EX.Currency))
            elif key == 'language':
                g.add( (URIRef(result["o"]["value"]), DBO.language, URIRef(value['value'])))
                g.add( (URIRef(value['value']), RDF.type, EX.Language))

serialize()
save('country_info.ttl')
```
